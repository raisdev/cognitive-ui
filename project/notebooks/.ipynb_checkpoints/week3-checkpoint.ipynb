```python
search = GoogleSerperAPIWrapper()
image_search = GoogleSerperAPIWrapper(type="images")
tool = Tool(
    name="google_search",
    description="Search Google for recent results.",
    func=search.run,
)
image_search_tool = Tool(
    name="google_image_search",
    description="Search Google for recent image results.",
    func=image_search.run,
)
    
class Collection(BaseModel):
    collection: str = Field(description="Answer to the input. Seperated by the newline token. Give the answer only.")

class block(BaseModel):
    title: str = Field(description="The title of the object.")
    descrption: str = Field(description="The description of the object")
    

def list_query(query: str) -> str:
    model = ChatOpenAI(model_name=gpt_version, temperature=0)
    results_found = search.run(query)
    parser = JsonOutputParser(pydantic_object=Collection)

    prompt = PromptTemplate(
        template="Answer the user query.\n{format_instructions}\n{query}\n. Use the following information found online {results_found}\n",
        input_variables=["query","results_found"],
        partial_variables={"format_instructions": parser.get_format_instructions()},
    )
    chain = prompt | model | parser
    result = chain.invoke({"query": query, "results_found": results_found})
    return result["collection"].split("\n")

parser = JsonOutputParser(pydantic_object=Collection)
list_prompt = PromptTemplate(
        template="Answer the user query.\n{format_instructions}\n{query}\n.",
        input_variables=["query"],
        partial_variables={"format_instructions": parser.get_format_instructions()},
)

llm_chain = LLMChain(llm=llm, prompt=list_prompt)
list_tool = Tool(
    name="List",
    func=llm_chain.run,
    description="Generate a list of object using query and previously found information. Use whenever user want a list of objects.",
)


#query = "Top 5 recommended Anime."
#results_found = search.run("Top 5 recommended Anime.")
#list_query(query, results_found)
def extract(query: str):
    model = ChatOpenAI(model_name=gpt_version, temperature=0)
    results_found = search.run(query)
    parser = JsonOutputParser(pydantic_object=block)
    prompt = PromptTemplate(
        template="Answer the user query.\n{format_instructions}\n{query}\n. Use the following information found online {results_found}\n",
        input_variables=["query","results_found"],
        partial_variables={"format_instructions": parser.get_format_instructions()},
    )
    chain = prompt | model | parser
    result = chain.invoke({"query": query, "results_found": results_found})
    return result
#query = "One Punch Man"
#extract(query)
block_parser = JsonOutputParser(pydantic_object=block)
extract_prompt = PromptTemplate(
    template="Extract information based on the query.\n{format_instructions}\n{query}\n.",
    input_variables=["query",],
    partial_variables={"format_instructions": block_parser.get_format_instructions()},
)

extract_chain = LLMChain(llm=llm, prompt=extract_prompt)
extract_tool = Tool(
    name="Extract",
    func=extract_chain.run,
    description="Extract information and output in a json format that improves user experience. Always use this tool.",
)
```
